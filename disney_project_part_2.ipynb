````python
def explore_data_with_none(data, start, end):
    """
    Prints rows between start and end (exclusive).
    If end is None, prints until the last row.
    """
    data_slice_list = []
    
    if end == "None":
        end = None
    else:
        end=int(end)
        
    if end is None:
        end = len(data)

    for i in range(start, end):
        data_slice_list.append(data[i])
            
    return data_slice_list
````

## Step 2c — Add the optional rows/columns count

```python
def explore_data(data_list, start, end, rows_and_columns=False):
```

  1. Write an **if-else** statement:
      - If **`end`** is `None`, slice data from **`start:`** and assign it to **`data_slice_list`**.
      - Else, slice data from **`start`** to **`end`** and assign it to **`data_slice_list`**.

  2. Write a **for loop** to iterate over each row in **`data_slice_list`**:
      - Print the row.
      - Print an empty line for readability.

  3. Write an **if statement**:
      - If **`rows_and_columns`** is `True`, print:
          - Total number of rows using `len()` function.
          - Total number of columns using `len()` function. Remember you are dealing with lists.


````python
def explore_data(data, start, end, rows_and_columns):
    """
    Prints rows between start and end (exclusive), 
    with optional total row/column count.

    Args:
        data_list (list): The dataset as a list of lists
        start (int): The index of the first row to print
        end (int or None): The index (exclusive) where we stop printing, or None to print to the end
        rows_and_columns (bool): Whether to display total row and column count. Default is False.

    Returns:
        None
    """
    data_slice_list = []
    
    if end == "None":
        end = None
    else:
        end=int(end)
        
    if end is None:
        end = len(data)

    for i in range(start, end):
        data_slice_list.append(data[i])
        
    if rows_and_columns:
        total_rows = len(data_slice_list)
        if total_rows > 0:
            total_columns = len(data_slice_list[0])
        else:
            total_columns = 0

        print("Total rows:", total_rows)
        print("Total columns:", total_columns)
        return data_slice_list, total_rows, total_columns
    
    else:
        return data_slice_list
````

Now that you’ve done it manually, let’s turn it into a function so you can reuse it for any column.

2. Write a for loop to iterate over each row element in **`data`**
    - If element in the row at **`col_index`** is not in the **`result_list`**, append it.

For example, I want to get unique values in `rating`column.
I will call this function and the output will be:

````python
# Update the function definition based on the Pseudocode above
def list_of_elements(data, col_index):
    """
    Extract unique values from a specified column in a 2D dataset.

    Args:
        data (list of list): The dataset as a list of rows, where each row is a list of values.
        col_index (int): The index of the column to extract unique values from.
        skip_header (bool, optional): Whether to skip the first row (commonly a header). Defaults to True.

    Returns:
        list: A list of unique values from the specified column, preserving the order of first occurrence.
    """
    result_list = []

    for i in range(1, len(data)):  
        element = data[i][col_index]
        if element not in result_list:
            result_list.append(element)
    return result_list
````

````python
def elements_count(data, col_index):
    """
    Get specific values from all rows in a specified column in a dataset.

    Args:
        data (list[list]): The dataset as a list of lists.
        col_index (int): The index of the column to extract unique values from.

    Returns:
        count: count result of the specific value from the specific column.
    """
    element_count_dict={}
    for i in range(len(data)):
        rating = data[i][col_index].upper()
        if rating in element_count_dict:
            element_count_dict[rating]+=1
        else:
            element_count_dict[rating]=1

    return element_count_dict
````

### Step 5d — Turn Into a Reusable Function `elements_count()`

Goal: Create a function that counts unique values in a chosen column and returns a sorted dictionary.

```python 
def elements_count(data, col_index):
```

1. Create an empty dictionary called `element_count_dict`.

2. Write a for loop to iterate over each row in `data`:

    - Get the value from the specified `col_index`.

    - If the value is not in `element_count_dict`, add it with a `count` of 1.

    - Otherwise, increment its `count` by 1.

3. Sort `element_count_dict` by values in `descending order`.

4. Return the sorted dictionary.


For example, I want to get unique values and corresponding count in **`rating`** column in **`disney_movies`**.
I will call this function on **`disney_movies`** and the output will be:

```python
{'G': 253, 'PG': 235, 'TV-G': 233, 'TV-PG': 181, 'PG-13': 66, 'TV-14': 37, 'TV-Y7': 36, 'TV-Y7-FV': 7, 'TV-Y': 3, '': 1}

````python
# Update the function definition based on the Pseudocode above
def elements_count(data, col_index):
    
    """
    counts unique values in a chosen column.

    Args:
        data (list[list]): The dataset as a list of lists.
        col_index (int): The index of the column to extract unique values from.

    Returns:
        sorted_elements_count: returns a sorted dictionary.
    """
    element_count_dict = {}

    for i in range(len(data)):
        rating = data[i][col_index].upper()
        if rating in element_count_dict:
            element_count_dict[rating]+=1
        else:
            element_count_dict[rating]=1

    sorted_elements_count=dict(sorted(element_count_dict.items(), key=lambda x: x[1], reverse=True)) # This will sort the above created dictionary based on the count values. Search for 'sorted python documentation' in Google and learn more on this.

    return sorted_elements_count
````

````python
def elements_count(data, col_index):
    """
    Counts the occurrences of each element in a specified column of a dataset and returns them sorted in descending order.

    Args:
        data (list of lists): The dataset, where each inner list represents a row.
        col_index (int): The index of the column to analyze. Each cell in this column can contain multiple items separated by commas.

    Returns:
        dict: A dictionary where the keys are unique elements from the specified column, and the values are their counts, sorted in descending order by count.
    """
    element_count_dict = {}

    for row in range(len(data)):
        splitted_text= data[row][col_index].split(',')
        for item in splitted_text:
            category = item.strip()
            if category in element_count_dict:
                element_count_dict[category]+=1
            else:
                element_count_dict[category]=1

    sorted_by_values_desc= sort_elements(element_count_dict)
    return sorted_by_values_desc
````

````python
# Method 2: when calling the filter_data function, check for all the data and already return the filtered data which include the rows that match the filter criteria

def filter_data(data, filter_local):
    ''' 
    This function takes the full data and filter criteria,
    and returns the filtered rows.
    '''
    filtered_data = []  # This will store matching rows

    for row_local in data:
        include_row_ll = True  # Assume row is valid

        for col_index, filter_value in filter_local.items():
            cell_value = row_local[col_index]

            if "," in str(cell_value):  # Comma-separated values
                cell_values_list = [v.strip() for v in cell_value.split(",")]
                if filter_value not in cell_values_list:
                    include_row_ll = False
                    break
            else:  # Single value
                if str(cell_value).strip() != str(filter_value).strip():
                    include_row_ll = False
                    break

        if include_row_ll:
            filtered_data.append(row_local)  # ✅ append matching row

    return filtered_data

def elements_count(data, col_index, sep=None, filters={}):
    filtered_data = filter_data(data, filters)
    element_count_dict = {}

    for row in filtered_data:
        cell_value = row[col_index]
        values = cell_value.split(sep) if sep else [cell_value]

        for val in values:
            val = val.strip()
            element_count_dict[val] = element_count_dict.get(val, 0) + 1

    sorted_elements_count = dict(
        sorted(element_count_dict.items(), key=lambda x: x[1], reverse=True)
    )

    return sorted_elements_count
````
